<h1>Analysis Type: Regional</h1>

```python
df['region_txt']
```

```python
filter1 = df[df['region_txt'] == 'Middle East & North Africa']
```

```python
df['attacktype1_txt'].value_counts()
```

```python
filter2 = filter1[filter1['attacktype1_txt'] == 'Bombing/Explosion']
```

```python
filter3 = filter2[filter2['nkill'] >= 10]
```

```python
df['iyear']
```

```python
filter4 = filter3[filter3['iyear'] >= 2000]
```

```python
import folium
from folium.plugins import HeatMap
```

```python
SE_terror = filter4.groupby(['longitude', 'latitude']).size().reset_index(name='incident_count')
SE_terror
```

```python
heat_df = SE_terror[['longitude', 'latitude', 'incident_count']].values.tolist()
heat_df
```

```python
my_map = folium.Map(location=[30, 40], zoom_start=4)
```

```python
HeatMap(heat_df).add_to(my_map)
my_map

```

<h1>Analysis Type: Crime Type</h1>

```python
southeast_asia_filter = df[(df['region_txt'] == 'Southeast Asia') & (df['attacktype1_txt'] == 'Armed Assault')]

```

```python
crime_data = southeast_asia_filter.groupby(['longitude', 'latitude']).size().reset_index(name='incident_count')

```

```python
heatmap_data_sea = crime_data[['latitude', 'longitude', 'incident_count']].values.tolist()

```

```python
crime_map_sea = folium.Map(location=[10, 115], zoom_start=5, tiles="CartoDB dark_matter")

```

```python
HeatMap(heatmap_data_sea, min_opacity=0.3, radius=12, blur=8, max_zoom=1).add_to(crime_map_sea)
crime_map_sea
```

<H1>-------------------------------------------------------------------------------------------</H1>

<h1>Analysis Type: Casualty</h1>

```python
ca_filter = df[df['region_txt'] == 'Central America & Caribbean']
```

```python
casualty_data = ca_filter.groupby(['longitude', 'latitude'])['nkill'].sum().reset_index(name='casualties')
```

```python
heatmap_data = casualty_data[['latitude', 'longitude', 'casualties']].values.tolist()
```

```python
map_ca = folium.Map(location=[15, -85], zoom_start=5, tiles="CartoDB dark_matter")
HeatMap(heatmap_data, min_opacity=0.3, radius=12, blur=8, max_zoom=1).add_to(map_ca)
map_ca
```

<h1>-------------------------------------------------------------------------------------------</h1>

<h1>Analysis Type: Demographic</h1>

```python
ca_filter = df[df['region_txt'] == 'Central America & Caribbean'].dropna(subset=['longitude', 'latitude', 'country_txt'])
casualty_data = ca_filter.groupby(['longitude', 'latitude', 'country_txt']).size().reset_index(name='incident_count')
heatmap_data = casualty_data[['latitude', 'longitude', 'incident_count']].values.tolist()
map_ca = folium.Map(location=[15, -85], zoom_start=5, tiles="CartoDB dark_matter")
HeatMap(heatmap_data, min_opacity=0.3, radius=10, blur=6).add_to(map_ca)
map_ca
```

<h1>Analysis Type: Target</h1>

<h2>focus on south east asia, government being target of bombing and explosion</h2>

```python
sea_ = df[(df['region_txt'] == 'Southeast Asia') & (df['targtype1_txt'] == 'Government (General)') & (df['attacktype1_txt'] == 'Bombing/Explosion')].dropna(subset=['longitude', 'latitude', 'country_txt'])
target = sea_.groupby(['longitude', 'latitude', 'country_txt']).size().reset_index(name='incident_count')

heatmap_data = target[['latitude', 'longitude', 'incident_count']].values.tolist()

map_sea = folium.Map(location=[10, 115], zoom_start=5, tiles="CartoDB dark_matter")

HeatMap(heatmap_data, min_opacity=0.3, radius=10, blur=6).add_to(map_sea)

map_sea
```
